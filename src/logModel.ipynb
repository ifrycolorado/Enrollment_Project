```python
import pandas as pd 
import numpy as np
from matplotlib import pyplot as plt
```

```python
# df = pd.read_csv('../data/CS Course Enrollment Data.csv')
df = pd.read_csv('/Users/evanwhite/Downloads/CS Course Enrollment Data.csv')
```

```python
# assign id to each row
df['id'] = range(1, len(df) + 1)
df
```

```python
# create list of valid semesters 
valid_semesters = [i + " " + j for i in ['Fall', 'Spring', 'Summer One', 'Summer Two', 'Summer'] for j in ['20' + str(year) for year in range(17, 24)]]

df = df[df.iloc[:, 3:-1].apply(lambda x: all(pd.isna(item) or item in valid_semesters for item in x), axis=1)]
df
```

```python
# pivot table from wide to long
melted_df = pd.melt(df, id_vars=['Semester Admitted', 'Department', 'Major', 'id'], var_name='Class_Semester', value_name='Taken')

# filter out NaN rows
melted_df = melted_df.dropna(subset=['Taken', 'Department', 'Major'])

# split 'Class_Semester' into 'Class' and 'Semester'
melted_df[['Class', 'Semester_Type']] = melted_df['Class_Semester'].str.split('_', n=1, expand=True)
melted_df = melted_df.drop(columns=['Class_Semester'])

# split Semester_Type into Pass or Tran
melted_df[['Type', 'Garbage']] = melted_df['Semester_Type'].str.split('_', n=1, expand=True)
melted_df = melted_df.drop(columns=['Garbage', 'Semester_Type'])

# split taken 
melted_df[['Taken_Semester', 'Taken_Year']] = melted_df['Taken'].str.rsplit(' ', n=1, expand=True)
melted_df[['Admitted_Semester', 'Admitted_Year']] = melted_df['Semester Admitted'].str.rsplit(' ', n=1, expand=True)

# replace 261 with 200, 262 with 220 
melted_df['Class'] = melted_df['Class'].replace('CSCI261', 'CSCI200')
melted_df['Class'] = melted_df['Class'].replace('CSCI262', 'CSCI220')

# Display the first few rows to verify
melted_df.sort_values(by='id').head(10)
```

```python
# remove messed up admit dates
melted_df = melted_df.astype({'Taken_Year': 'int32', 'Admitted_Year': 'int32'})

UNIQUE_SEM = 4
melted_df['Taken_Num'] = np.select([
    melted_df['Taken_Semester'] == 'Spring',
    melted_df['Taken_Semester'] == 'Summer One',
    melted_df['Taken_Semester'] == 'Summer Two',
    melted_df['Taken_Semester'] == 'Fall'
    ], 
    [melted_df['Taken_Year'] * UNIQUE_SEM, 
    melted_df['Taken_Year'] * UNIQUE_SEM + 1, 
    melted_df['Taken_Year'] * UNIQUE_SEM + 2,
    melted_df['Taken_Year'] * UNIQUE_SEM + 3], 
    default = 'UNKNOWN')

melted_df['Admitted_Num'] = np.select([
    melted_df['Admitted_Semester'] == 'Spring',
    melted_df['Admitted_Semester'] == 'Summer One',
    melted_df['Admitted_Semester'] == 'Summer Two',
    melted_df['Admitted_Semester'] == 'Summer',
    melted_df['Admitted_Semester'] == 'Fall'
    ], 
    [melted_df['Admitted_Year'] * UNIQUE_SEM, 
    melted_df['Admitted_Year'] * UNIQUE_SEM + 1, 
    melted_df['Admitted_Year'] * UNIQUE_SEM + 2,
    melted_df['Admitted_Year'] * UNIQUE_SEM + 1,
    melted_df['Admitted_Year'] * UNIQUE_SEM + 3], 
    default = 'UNKNOWN')
```

```python
# remove invalid admit dates
invalids = melted_df[melted_df['Taken_Num'] < melted_df['Admitted_Num']]
invalid_ids = invalids['id'].unique()
melted_df = melted_df[~melted_df['id'].isin(invalid_ids)]
```

```python
master_df = log_df.join(df, on='id', rsuffix='_other', how = 'left')
targetCourse = 406

del(master_df['Major'])
del(master_df['Department'])

master_df['CSCI200_Pass_Sem'] = master_df['CSCI200_Pass_C_Sem']             # Get rid of _C_
master_df['CSCI261_Pass_Sem'] = master_df['CSCI261_Pass_C_Sem']

master_df[['Admitted_Semester', 'Admitted_Year']] = master_df['Semester Admitted'].str.rsplit(' ', n=1, expand=True)
del(master_df['Semester Admitted'])
master_df[[f'CSCI{targetCourse}_Taken_Semester', f'CSCI{targetCourse}_Taken_Year']] = master_df['Semester'].str.rsplit(' ', n=1, expand=True)
master_df[f'CSCI{targetCourse}_Taken_Year'] = master_df[f'CSCI{targetCourse}_Taken_Year'].astype('Int64')
master_df['Admitted_Year'] = master_df['Admitted_Year'].astype('Int64')

for course in [101, 128, 200, 220, 261, 262]:
    master_df[f'CSCI{course}_Pass_Sem'].fillna(master_df[f'CSCI{course}_Tran_Sem'], inplace=True)       # Merge pass and transfer cols for each class
    del(master_df[f'CSCI{course}_Tran_Sem'])

master_df['CSCI128_Pass_Sem'].fillna(master_df['CSCI101_Pass_Sem'], inplace=True)       # Account for course number changes
master_df['CSCI200_Pass_Sem'].fillna(master_df['CSCI261_Pass_Sem'], inplace=True)
master_df['CSCI220_Pass_Sem'].fillna(master_df['CSCI262_Pass_Sem'], inplace=True)

for var in ['unique_id', 'Class', 'id_other', 'CSCI101_Pass_Sem', 'CSCI200_Pass_C_Sem', 'CSCI261_Pass_Sem', 'CSCI261_Pass_C_Sem',     # Drop cols we dont need
            'CSCI262_Pass_Sem', f'CSCI{targetCourse}_Pass_Sem', 'CSCI210_Pass_Sem']:
    del(master_df[var])

for course in [128, 200, 220, 274, 306, 341, 358, 370, 400, 442]:
    master_df[[f'CSCI{course}_Taken_Semester', f'CSCI{course}_Taken_Year']] = master_df[f'CSCI{course}_Pass_Sem'].str.rsplit(' ', n=1, expand=True)
    master_df[f'CSCI{course}_Taken_Year'] = master_df[f'CSCI{course}_Taken_Year'].astype('Int64')
    master_df[f'CSCI{course}_Taken_Year'] = master_df[f'CSCI{course}_Taken_Year'].fillna(3000)
    del(master_df[f'CSCI{course}_Pass_Sem'])

master_df.to_csv('/Users/evanwhite/Downloads/CS_Course_Enrollment_Data_Master.csv', index=False)
master_df = master_df[master_df['Admitted_Semester'].notna()]
master_df.head()



```

```python
for course in [128, 200, 220, 274, 306, 341, 358, 370, 400, 442]:
    master_df[f'CSCI{course}'] = 0
master_df[f'AcademicYear'] = 0

def precedence(row, course):
    takenSemester = row[f'CSCI{targetCourse}_Taken_Semester']
    takenYear = (row[f'CSCI{targetCourse}_Taken_Year'])

    if row[f'CSCI{course}_Taken_Year'] < takenYear: return 1
    if row[f'CSCI{course}_Taken_Year'] == takenYear and row[f'CSCI{course}_Taken_Semester'] != takenSemester:
        if row[f'CSCI{course}_Taken_Semester'] == 'Spring': return 1
        if takenSemester == 'Fall' and row[f'CSCI{course}_Taken_Semester'] != 'Fall': return 1
        if row[f'CSCI{course}_Taken_Semester'] == 'Spring' and takenSemester in ['Summer', 'Summer One', 'Summer Two']: return 1
    return 0

def academicYear(row):
    takenSemester = row[f'CSCI{targetCourse}_Taken_Semester']
    takenYear = (row[f'CSCI{targetCourse}_Taken_Year'])

    admittedSemester = row['Admitted_Semester']
    admittedYear = row['Admitted_Year']

    if admittedYear > takenYear: return 0
    if admittedYear == takenYear:
        if admittedSemester == takenSemester: return 1
        if admittedSemester in ['Summer', 'Summer One', 'Summer Two'] and takenSemester in ['Summer', 'Summer One', 'Summer Two']: return 1
        if admittedSemester == 'Spring': return 1
        if admittedSemester in ['Summer', 'Summer One', 'Summer Two']:
            if takenSemester == 'Spring': return 0
            else: return 1
        if admittedSemester == 'Fall': return 0
    else:
        if admittedSemester == takenSemester: return takenYear - admittedYear + 1
        if admittedSemester in ['Summer', 'Summer One', 'Summer Two'] and takenSemester in ['Summer', 'Summer One', 'Summer Two']: return takenYear - admittedYear + 1
        else:
            if admittedSemester == 'Fall': return takenYear - admittedYear
            if admittedSemester == 'Spring': return takenYear - admittedYear + 1
            else: 
                if takenSemester == 'Fall': return takenYear - admittedYear + 1
                if takenSemester == 'Spring': return takenYear - admittedYear
            

for index, row in master_df.iterrows():
    for course in [128, 200, 220, 274, 306, 341, 358, 370, 400, 442]:
        master_df.at[index, f'CSCI{course}'] = int(precedence(row, course))
        master_df.at[index, 'AcademicYear'] = int(academicYear(row))

master_df
```
